```python
categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)
numerical_columns = list(df.dtypes[df.dtypes != 'object'].index)
numerical_columns
```

```python
df[categorical_columns] = df[categorical_columns].fillna('NA')
df[numerical_columns] = df[numerical_columns].fillna(0.0)
df.isnull().sum()
```

```python
df_corr_full_train = df
df_corr_converted = df['converted']
del df_corr_full_train['converted']
df_corr_converted
```

```python
from sklearn.metrics import mutual_info_score
corr_numerical_columns = numerical_columns
#corr_numerical_columns.remove('converted')
corr_numerical_columns
```

```python
df_corr_full_train[numerical_columns].corrwith(df_corr_converted).abs()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
df
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.converted.values
y_val = df_val.converted.values
y_test = df_test.converted.values

del df_train['converted']
del df_val['converted']
del df_test['converted']
```

```python
def mutual_info_converted_score(series):
  return mutual_info_score(series, y_train)
```

```python
mi = df_train[categorical_columns].apply(mutual_info_converted_score)
round(mi.sort_values(ascending=False), 2)
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical_columns + corr_numerical_columns].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical_columns + corr_numerical_columns].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
convert_choice = (y_pred >= 0.5)
```

```python
original_accuracy = (y_val == convert_choice).mean()
```

```python
col_stack = categorical_columns + corr_numerical_columns
```

```python
def create_omitted_dictionary(original_list):
    omitted_dict = {}
    for i in range(len(original_list)):
        omitted_item = original_list[i]
        list_without_omitted = original_list[:i] + original_list[i+1:]
        omitted_dict[omitted_item] = list_without_omitted

    return omitted_dict
```

```python
columns_explode = create_omitted_dictionary(col_stack)
columns_explode
```

```python
for omitted_col, new_cols in columns_explode.items():
  train_multiple(omitted_col, new_cols)
```

```python
def train_single(regularizer):
  t_dict = df_train[col_stack].to_dict(orient='records')
  Xi_train = dv.fit_transform(t_dict)

  v_dict = df_val[col_stack].to_dict(orient='records')
  Xi_val = dv.transform(v_dict)

  model = LogisticRegression(solver='liblinear', C=regularizer, max_iter=1000, random_state=42)
  model.fit(Xi_train, y_train)
  yy_pred = model.predict_proba(Xi_val)[:, 1]
  convert_choice = (yy_pred >= 0.5)
  accuracy = (y_val == convert_choice).mean()
  print(f"{regularizer} : {accuracy} : {abs(original_accuracy - accuracy)}")
```
